```python
def print_keys(data):
    if isinstance(data, dict):
        for key in data:
            print(f"Key: {key}")
            # Optionally print the value for each key
            # print(f"Value: {data[key]}")
    elif isinstance(data, list):
        for item in data:
            print_keys(item)  # Recursive call for list items

# Print keys
print_keys(json_data)
```
